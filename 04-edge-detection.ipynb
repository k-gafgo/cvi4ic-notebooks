Detecting edges using gradient-based methods.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Test Image

```python
# Create a test image with geometric shapes
img = np.zeros((300, 300), dtype=np.uint8)
cv2.rectangle(img, (50, 50), (150, 150), 200, -1)
cv2.circle(img, (220, 220), 50, 150, -1)
cv2.ellipse(img, (150, 80), (40, 20), 30, 0, 360, 180, -1)

# Add some Gaussian blur to make gradients smoother
img = cv2.GaussianBlur(img, (5, 5), 1.0)

plt.imshow(img, cmap="gray")
plt.title("Test Image")
plt.axis("off")
plt.show()
```

## 2. Sobel Operator

```python
# Compute gradients
grad_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Gradient magnitude and direction
magnitude = np.sqrt(grad_x**2 + grad_y**2)
direction = np.arctan2(grad_y, grad_x)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].imshow(grad_x, cmap="RdBu")
axes[0, 0].set_title("Gradient X (Sobel)")
axes[0, 1].imshow(grad_y, cmap="RdBu")
axes[0, 1].set_title("Gradient Y (Sobel)")
axes[1, 0].imshow(magnitude, cmap="gray")
axes[1, 0].set_title("Gradient Magnitude")
axes[1, 1].imshow(direction, cmap="hsv")
axes[1, 1].set_title("Gradient Direction")
for ax in axes.flat:
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 3. Canny Edge Detection

```python
# Canny with different thresholds
edges_low = cv2.Canny(img, 30, 90)
edges_mid = cv2.Canny(img, 50, 150)
edges_high = cv2.Canny(img, 100, 200)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, edges, title in zip(
    axes,
    [edges_low, edges_mid, edges_high],
    ["Canny (30, 90)", "Canny (50, 150)", "Canny (100, 200)"]
):
    ax.imshow(edges, cmap="gray")
    ax.set_title(title)
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 4. Laplacian

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)

plt.figure(figsize=(8, 6))
plt.imshow(np.abs(laplacian), cmap="gray")
plt.title("Laplacian (absolute)")
plt.axis("off")
plt.colorbar()
plt.show()
```
