````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print(f"OpenCV version: {cv2.__version__}")
print(f"NumPy version: {np.__version__}")
````

## 2. Loading and Displaying Images

````python
# Load an image (replace with your own image path)
img = cv2.imread("sample.jpg")

# Check if loading was successful
if img is None:
    # Create a sample image if no file is available
    img = np.random.randint(0, 255, (256, 256, 3), dtype=np.uint8)
    print("Using a random sample image")

print(f"Image shape: {img.shape}")
print(f"Image dtype: {img.dtype}")
print(f"Image size: {img.size} pixels")
````

````python
# Display with matplotlib (convert BGR to RGB)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 6))
plt.imshow(img_rgb)
plt.title("Sample Image")
plt.axis("off")
plt.show()
````

## 3. Basic Image Operations

````python
# Access individual pixel values
pixel = img[100, 100]
print(f"Pixel at (100, 100): B={pixel[0]}, G={pixel[1]}, R={pixel[2]}")

# Crop a region
crop = img[50:150, 50:200]
print(f"Cropped region shape: {crop.shape}")

# Resize
resized = cv2.resize(img, (128, 128))
print(f"Resized shape: {resized.shape}")
````

## 4. Drawing on Images

````python
# Create a canvas
canvas = np.zeros((300, 400, 3), dtype=np.uint8)

# Draw shapes
cv2.rectangle(canvas, (50, 50), (150, 150), (0, 255, 0), 2)
cv2.circle(canvas, (300, 100), 50, (0, 0, 255), -1)
cv2.line(canvas, (0, 250), (400, 250), (255, 255, 0), 3)
cv2.putText(canvas, "Hello CV!", (100, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

plt.figure(figsize=(8, 5))
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.title("Drawing on Images")
plt.axis("off")
plt.show()
````
