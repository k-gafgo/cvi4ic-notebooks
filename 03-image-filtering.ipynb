Applying spatial filters to images using convolution.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Create a Noisy Test Image

```python
# Create a simple test image with shapes
img = np.zeros((256, 256), dtype=np.uint8)
cv2.rectangle(img, (50, 50), (200, 200), 200, -1)
cv2.circle(img, (128, 128), 40, 100, -1)

# Add Gaussian noise
noise = np.random.normal(0, 25, img.shape).astype(np.float64)
noisy = np.clip(img.astype(np.float64) + noise, 0, 255).astype(np.uint8)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(img, cmap="gray")
axes[0].set_title("Original")
axes[1].imshow(noisy, cmap="gray")
axes[1].set_title("Noisy")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 2. Smoothing Filters

```python
# Apply different smoothing filters
box = cv2.blur(noisy, (5, 5))
gaussian = cv2.GaussianBlur(noisy, (5, 5), sigmaX=1.0)
median = cv2.medianBlur(noisy, 5)
bilateral = cv2.bilateralFilter(noisy, 9, 75, 75)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
for ax, filtered, title in zip(
    axes.flat,
    [box, gaussian, median, bilateral],
    ["Box Filter (5x5)", "Gaussian (5x5, σ=1)", "Median (5)", "Bilateral"]
):
    ax.imshow(filtered, cmap="gray")
    ax.set_title(title)
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 3. Custom Kernels

```python
# Sharpening kernel
sharpen_kernel = np.array([[ 0, -1,  0],
                           [-1,  5, -1],
                           [ 0, -1,  0]], dtype=np.float32)

sharpened = cv2.filter2D(noisy, -1, sharpen_kernel)

# Emboss kernel
emboss_kernel = np.array([[-2, -1,  0],
                          [-1,  1,  1],
                          [ 0,  1,  2]], dtype=np.float32)

embossed = cv2.filter2D(img, -1, emboss_kernel)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(sharpened, cmap="gray")
axes[0].set_title("Sharpened")
axes[1].imshow(embossed, cmap="gray")
axes[1].set_title("Embossed")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 4. Effect of Kernel Size

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for ax, ksize in zip(axes, [3, 7, 15, 31]):
    blurred = cv2.GaussianBlur(noisy, (ksize, ksize), 0)
    ax.imshow(blurred, cmap="gray")
    ax.set_title(f"Gaussian {ksize}x{ksize}")
    ax.axis("off")
plt.tight_layout()
plt.show()
```
