Exploring how images are stored as arrays and working with different color spaces.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Image as a NumPy Array

```python
# Create a synthetic test image
img = np.zeros((200, 300, 3), dtype=np.uint8)
img[0:100, 0:150] = [255, 0, 0]    # Blue (BGR)
img[0:100, 150:300] = [0, 255, 0]   # Green
img[100:200, 0:150] = [0, 0, 255]   # Red
img[100:200, 150:300] = [255, 255, 0]  # Cyan

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Test Image (BGR -> RGB)")
plt.axis("off")
plt.show()
```

## 2. Color Channels

```python
# Split into individual channels
b, g, r = cv2.split(img)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, channel, title in zip(axes, [r, g, b], ["Red", "Green", "Blue"]):
    ax.imshow(channel, cmap="gray")
    ax.set_title(f"{title} Channel")
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 3. Color Space Conversions

```python
# Convert to different color spaces
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].imshow(gray, cmap="gray")
axes[0].set_title("Grayscale")
axes[1].imshow(hsv)
axes[1].set_title("HSV (raw channels)")
axes[2].imshow(lab)
axes[2].set_title("Lab (raw channels)")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
plt.show()
```

## 4. HSV Color Segmentation

```python
# Segment a specific color range in HSV
lower = np.array([100, 50, 50])
upper = np.array([130, 255, 255])

mask = cv2.inRange(hsv, lower, upper)
result = cv2.bitwise_and(img, img, mask=mask)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original")
axes[1].imshow(mask, cmap="gray")
axes[1].set_title("Mask")
axes[2].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
axes[2].set_title("Segmented")
for ax in axes:
    ax.axis("off")
plt.tight_layout()
plt.show()
```
